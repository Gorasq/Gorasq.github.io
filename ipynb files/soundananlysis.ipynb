```python
import librosa
import matplotlib.pyplot as plt
```

```python
path = '/content/testaudio.m4a'
```

```python
librosa.Audio
```

```python
y, sr = librosa.load(path, sr=22050, offset = 0.0)
```

```python
print('y의 길이:', len(y))
print('y의 타입:', type(y))
print('sr = ',sr)
```

```python
img = librosa.display.waveshow(y = y, sr = sr, axis ='time')
```

```python
import numpy as np
```

```python
stft = librosa.stft(y)
librosa.display.specshow(librosa.amplitude_to_db(stft,ref=np.max), y_axis = 'log',x_axis = 'time',sr=sr)
```

```python
mfcc = librosa.feature.mfcc(y=y,sr=sr)
librosa.display.specshow(mfcc, x_axis='time', sr=sr)
```

```python
tempo, _ = librosa.beat.beat_track(y=y,sr=sr)
tempogram = librosa.feature.tempogram(y=y, sr=sr)
librosa.display.specshow(tempogram, x_axis = 'time', sr = sr)
```
